I. Machine Learning Algorithms

8. K-Means Clustering

- Mô tả: Phân nhóm dữ liệu vào các cụm dựa trên khoảng cách.
- Demo: Phân nhóm khách hàng dựa trên hành vi mua hàng.

```python
import numpy as np # Thư viện hỗ trợ tính toán
import pandas as pd # Thư viện hỗ trợ xử lý dữ liệu dạng bảng
from sklearn.cluster import KMeans # Thư viện hỗ trợ phân cụm
import matplotlib.pyplot as plt # Thư viện hỗ trợ vẽ đồ thị
import seaborn as sns # Thư viện hỗ trợ vẽ đồ thị

# Tạo dữ liệu giả định
np.random.seed(42) # Để kết quả giữ nguyên qua các lần chạy
n_samples = 200 # Số lượng mẫu dữ liệu
frequency = np.random.randint(1, 10, n_samples)  # Tần suất mua hàng
avg_order_value = np.random.randint(10, 100, n_samples)  # Giá trị trung bình của đơn hàng
num_products = np.random.randint(1, 5, n_samples)  # Số lượng sản phẩm trong giỏ

# Tạo DataFrame
data = pd.DataFrame({ # Tạo DataFrame từ dict dữ liệu
    'Frequency': frequency, # Tần suất mua hàng
    'AvgOrderValue': avg_order_value, # Giá trị trung bình của đơn hàng
    'NumProducts': num_products # Số lượng sản phẩm trong giỏ
})

# Sử dụng KMeans để phân cụm (giả sử 3 cụm)
kmeans = KMeans(n_clusters=3, random_state=42) # Khởi tạo mô hình KMeans với 3 cụm
data['Cluster'] = kmeans.fit_predict(data) # Gán nhãn cụm cho từng mẫu dữ liệu

# Hiển thị các trung tâm cụm
centers = kmeans.cluster_centers_ # Trung tâm của các cụm (3, 2)

# Vẽ biểu đồ phân cụm
plt.figure(figsize=(10, 6)) # Kích thước biểu đồ
sns.scatterplot(data=data, x='Frequency', y='AvgOrderValue', hue='Cluster', palette='viridis', s=100) # Vẽ biểu đồ phân cụm
plt.scatter(centers[:, 0], centers[:, 1], s=300, c='red', label='Cluster Centers') # Vẽ trung tâm cụm
plt.title("Phân cụm khách hàng dựa trên hành vi mua hàng") # Tiêu đề biểu đồ
plt.xlabel("Tần suất mua hàng") # Tên trục tung
plt.ylabel("Giá trị trung bình của đơn hàng") # Tên trục hoành
plt.legend() # Hiển thị chú thích
plt.show() # Hiển thị biểu đồ
```
