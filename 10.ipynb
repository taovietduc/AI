10. Principal Component Analysis (PCA)

- Mô tả: Giảm số chiều của dữ liệu trong khi bảo tồn thông tin quan trọng.
- Demo: Giảm chiều dữ liệu hình ảnh để tăng hiệu quả xử lý.

```python
import matplotlib.pyplot as plt # Thư viện vẽ đồ thị matplotlib (cần cài đặt: pip install matplotlib)
from sklearn.decomposition import PCA # Thư viện giảm chiều dữ liệu PCA của scikit-learn
from sklearn.datasets import load_digits # Dữ liệu hình ảnh chữ viết tay (0-9)
import seaborn as sns # Thư viện vẽ đồ thị đẹp mắt (cần cài đặt: pip install seaborn)

# Tải dữ liệu hình ảnh chữ viết tay (0-9)
digits = load_digits() # Dữ liệu gồm 1797 ảnh, mỗi ảnh có kích thước 8x8 pixels
data = digits.data # 1797 ảnh được biểu diễn dưới dạng ma trận 64 chiều (8x8)
target = digits.target # Nhãn của ảnh (0-9) tương ứng với mỗi ảnh

# Áp dụng PCA để giảm chiều từ 64 xuống còn 2
pca = PCA(n_components=2) # Giảm xuống còn 2 chiều (PC1, PC2) để vẽ đồ thị
data_pca = pca.fit_transform(data) # Thực hiện giảm chiều dữ liệu và chuyển đổi dữ liệu
 
# Chuyển kết quả thành DataFrame để trực quan hóa
import pandas as pd # Thư viện xử lý dữ liệu dạng bảng (cần cài đặt: pip install pandas)
df = pd.DataFrame(data_pca, columns=['PC1', 'PC2']) # Chuyển kết quả thành DataFrame
df['Target'] = target # Thêm cột nhãn (target) vào DataFrame

# Vẽ biểu đồ phân tán của dữ liệu sau khi giảm chiều
plt.figure(figsize=(10, 8)) # Tạo kích thước biểu đồ rộng 10 inch, cao 8 inch
sns.scatterplot(data=df, x='PC1', y='PC2', hue='Target', palette='tab10', s=60, legend='full') # Vẽ biểu đồ phân tán với 10 màu khác nhau
plt.title("PCA of Handwritten Digits Dataset (2 Components)") # Đặt tiêu đề biểu đồ là "PCA of Handwritten Digits Dataset (2 Components)"
plt.xlabel("Principal Component 1") # Đặt tên trục hoành là "Principal Component 1" (PC1)   
plt.ylabel("Principal Component 2") # Đặt tên trục tung là "Principal Component 2" (PC2)
plt.show() # Hiển thị biểu đồ vừa vẽ 
```
