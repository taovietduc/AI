I. Machine Learning Algorithms

9. Hierarchical Clustering

- Mô tả: Phân nhóm dữ liệu thành các cấp độ khác nhau.
- Demo: Phân loại giống hoa dựa trên các đặc điểm.

```python
import pandas as pd # Thư viện xử lý dữ liệu dạng bảng  (DataFrame)
from sklearn.preprocessing import StandardScaler  # Thư viện chuẩn hóa dữ liệu
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster # Thư viện hierarchical clustering
import matplotlib.pyplot as plt # Thư viện vẽ đồ thị
import seaborn as sns   # Thư viện vẽ đồ thị
from sklearn.datasets import load_iris # Dữ liệu Iris dataset từ thư viện sklearn

# Tải dữ liệu Iris
iris = load_iris() # Dữ liệu Iris dataset từ thư viện sklearn
data = pd.DataFrame(iris.data, columns=iris.feature_names) # Chuyển dữ liệu Iris thành DataFrame

# Chuẩn hóa dữ liệu
scaler = StandardScaler() # Khởi tạo đối tượng chuẩn hóa dữ liệu
data_scaled = scaler.fit_transform(data) # Chuẩn hóa dữ liệu Iris dataset theo z-score

# Áp dụng hierarchical clustering
Z = linkage(data_scaled, method='ward') # Tạo ma trận liên kết dùng phương pháp Ward

# Vẽ biểu đồ Dendrogram 
plt.figure(figsize=(12, 8)) # Kích thước biểu đồ Dendrogram
dendrogram(Z, labels=iris.target, leaf_rotation=90, leaf_font_size=10) # Vẽ Dendrogram
plt.title("Dendrogram for Hierarchical Clustering of Iris Dataset") # Tiêu đề biểu đồ
plt.xlabel("Sample Index") # Tên trục x 
plt.ylabel("Distance") # Tên trục y
plt.show() # Hiển thị biểu đồ Dendrogram

# Phân cụm thành 3 cụm dựa trên ngưỡng
clusters = fcluster(Z, t=3, criterion='maxclust') # Phân cụm dữ liệu thành 3 cụm dựa trên ngưỡng
data['Cluster'] = clusters # Thêm cột Cluster vào dữ liệu

# Vẽ phân cụm
sns.pairplot(data, hue='Cluster', palette='viridis', markers=["o", "s", "D"], diag_kind='kde') # Vẽ phân cụm
plt.suptitle("Hierarchical Clustering on Iris Dataset", y=1.02) # Tiêu đề biểu đồ phân cụm 
plt.show() # Hiển thị biểu đồ phân cụm 
```
