I. Machine Learning Algorithms
1. Linear Regression


```python
import numpy as np # Thư viện hỗ trợ tính toán trên mảng nhiều chiều
import matplotlib.pyplot as plt # Thư viện hỗ trợ vẽ đồ thị
from sklearn.linear_model import LinearRegression # Thư viện hỗ trợ mô hình hóa dữ liệu

# Dữ liệu diện tích (m²) và giá nhà (triệu VND)
area = np.array([50, 80, 100, 150, 200]).reshape(-1, 1)  # reshape để tạo cột cho dữ liệu
price = np.array([1.2, 2.0, 2.5, 3.8, 5.0]) # không cần reshape vì dữ liệu này đã là cột

# Tạo mô hình hồi quy tuyến tính
model = LinearRegression() # model = LinearRegression(fit_intercept=True) là mặc định nên có thể bỏ qua
model.fit(area, price) # Huấn luyện mô hình với dữ liệu

# Hệ số hồi quy và hằng số
beta_1 = model.coef_[0] # hệ số hồi quy (slope)
beta_0 = model.intercept_ # hằng số (intercept) hay bias

# Dự đoán giá nhà với diện tích 120m²
area_to_predict = np.array([[120]]) # Chú ý: cần reshape để tạo cột cho dữ liệu
predicted_price = model.predict(area_to_predict) # Dự đoán giá nhà

# Vẽ biểu đồ
plt.scatter(area, price, color='blue', label='Dữ liệu thực tế') # Vẽ dữ liệu thực tế (màu xanh)
plt.plot(area, model.predict(area), color='red', label=f'Đường hồi quy: y = {beta_1:.3f}x + {beta_0:.3f}') # Vẽ đường hồi quy (màu đỏ)
plt.scatter(area_to_predict, predicted_price, color='green', label=f'Dự đoán (120m²): {predicted_price[0]:.2f} triệu VND') # Vẽ dự đoán (màu xanh lá)

# Gán tiêu đề và chú thích
plt.title('Dự đoán giá nhà dựa trên diện tích') # Tiêu đề biểu đồ
plt.xlabel('Diện tích (m²)') # Tên trục hoành
plt.ylabel('Giá nhà (triệu VND)') # Tên trục tung
plt.legend()  # Chú thích

# Hiển thị kết quả
plt.show() # Hiển thị biểu đồ

# In ra kết quả dự đoán
print(f"Dự đoán giá nhà cho diện tích 120m²: {predicted_price[0]:.2f} triệu VND") # In ra giá nhà dự đoán
```
